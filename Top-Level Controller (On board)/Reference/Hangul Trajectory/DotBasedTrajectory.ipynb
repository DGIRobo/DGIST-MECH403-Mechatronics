Hangul Trajectory Showing Function

```python
import math
import matplotlib.pyplot as plt

def shower(word):
    x = []
    y = []
    z = []
    for dot in word:
        x.append(dot[0])
        y.append(dot[1])
        z.append(dot[2])
    ax = plt.figure(figsize=(5,5)).add_subplot(projection='3d')
    ax.axis([0, 1, 1, 0])
    ax.plot(x, y, z)
    ax.set_title("word trajectory 3d plot")

def stringShower(wordnum, string):
    x = []
    y = []
    z = []
    for dot in string:
        x.append(dot[0])
        y.append(dot[1])
        z.append(dot[2])
    ax = plt.figure().add_subplot(projection='3d')
    ax.axis([0, 0.7*wordnum, 0.7*wordnum, 0])
    ax.plot(x, y, z)
    ax.set_title("string trajectory 3d plot")
```

Word offset & resize, String maker

```python
def offset(word, x, y):
    word = word.copy()
    for dot in range(0,len(word)):
        word[dot] = list(word[dot])
    for dot in range(0, len(word)):
        word[dot][0] = word[dot][0] + x
        word[dot][1] = word[dot][1] + y
    for dot in range(0,len(word)):
        word[dot] = tuple(word[dot])
    return word

def resize(word, xscale, yscale):
    word = word.copy()
    for dot in range(0,len(word)):
        word[dot] = list(word[dot])
    for dot in range(0, len(word)):
        word[dot][0] = word[dot][0] * xscale
        word[dot][1] = word[dot][1] * yscale
    for dot in range(0,len(word)):
        word[dot] = tuple(word[dot])
    return word

def string_dot(size, chars):
    string_dot_list = []
    i = 0
    for char in chars:
        char = offset(char, i*size, 0)
        string_dot_list = string_dot_list + char
        i = i + 0.7
    return string_dot_list
```

```python
def Type1(size, word1, word2):
    word1 = resize(word1, size*55/100, size*4/5)
    word2 = resize(word2, size*35/100, size*1)
    word1 = offset(word1, 0, size*1/10)
    word2 = offset(word2, size*65/100, 0)
    return resize(word1+word2, 0.6, 1)

def Type2(size, word1, word2, word3):
    temp = Type1(size, word1, word2)
    temp = resize(temp, size, size*55/100)
    word3 = resize(word3, size*0.6, size*35/100)
    word3 = offset(word3, 0, size*65/100)
    return resize(temp+word3, 1, 1)

def Type3(size, word1, word2):
    word1 = resize(word1, size*4/5, size*55/100)
    word2 = resize(word2, size*1, size*35/100)
    word1 = offset(word1, size*1/10, 0)
    word2 = offset(word2, 0, size*65/100)
    return resize(word1+word2, 0.6, 1)

def Type4(size, word1, word2, word3):
    temp = Type3(size, word1, word2)
    temp = resize(temp, size, size*65/100)
    word3 = resize(word3, size*4/5*0.6, size*25/100)
    word3 = offset(word3, size*1/10, size*75/100)
    return resize(temp+word3, 1, 1)

def Type5(size, word1, word2):
    pass

def Type6(size, word1, word2):
    pass
```
